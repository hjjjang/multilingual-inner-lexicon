```python
import babelnet as bn
from babelnet import Language

synsets = bn.get_synsets("home", from_langs=[Language.EN])

for synset in synsets:
    print(f"Synset ID: {synset.id}")

    # Get German senses for the synset
    senses_de = bn.get_senses_from(synset.id, Language.DE)
    print("German translations:")
    for sense in senses_de:
        print(f" - {sense.lemma}")

    # Get Korean senses for the synset
    senses_ko = bn.get_senses_from(synset.id, Language.KO)
    print("Korean translations:")
    for sense in senses_ko:
        print(f" - {sense.lemma}")

    print("---------------------------")
```
